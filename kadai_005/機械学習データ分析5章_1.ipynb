```python
import matplotlib.pyplot as plt
import numpy as np

# サブプロットを作成
fig, axes = plt.subplots(nrows=2, ncols=3, tight_layout=True)

# 折れ線グラフ
y0 = [1, 2, -5, 2]
axes[0, 0].plot(y0)
axes[0, 0].set_title('Line Plot')

# sin関数
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
axes[0, 1].plot(x, y1)
axes[0, 1].set_title('Sine Function')

# ヒストグラム
data = np.random.randn(1000)
axes[0, 2].hist(data, bins=30)
axes[0, 2].set_title('Histogram')

# 散布図
x2 = np.random.rand(100)
y2 = np.random.rand(100)
axes[1, 0].scatter(x2, y2)
axes[1, 0].set_title('Scatter Plot')

# プロットされていない領域にタイトルを付ける
for ax in axes.flat:
    if not ax.has_data():
        ax.set_title('Empty Plot')


# グラフを表示
plt.show()
```
